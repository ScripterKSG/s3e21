```python
# drop outliers based on data from describe and our scatterplot
df = df.drop(df[df['target'] > 20].index)
df = df.drop(df[df['O2_1'] > 20].index)
df = df.drop(df[df['O2_2'] > 20].index)
df = df.drop(df[df['O2_3'] > 20].index)
df = df.drop(df[df['O2_4'] > 20].index)
df = df.drop(df[df['O2_5'] > 20].index)
df = df.drop(df[df['O2_6'] > 20].index)
```

```python
# pick features based on highest correalations
top10 = df.corr()[['target']].abs().unstack().sort_values(ascending=False).drop_duplicates().nlargest(10)
top10.index
```

```python
chosen_feat = ['id', 'target', 'O2_1', 'O2_2', 'O2_6', 'NH4_5', 'O2_3', 'O2_7', 'NH4_4', 'NO3_4', 'BOD5_2']
chosen_feat
```

```python
# this competition does not let us drop unwanted columns, so we set unwanted column vals all to 0
for col in df:
    if col not in chosen_feat:
        df[col] = 0
df
```

```python
# clip rest of our data to be within reasonable bounds
ids = df['id']
df = df.drop(['id'], axis = 1)
df = df.clip(0,16)
df.insert(0, 'id', ids)
df
```
